```python
with open("output.tmp", "rb") as f:
    data = list(f.read())
    print(data)
```

```python
def find_preamble_all(tx, preamble):
    preamble_len = len(preamble)
    indices = []  
    for i in range(len(tx) - preamble_len + 1):
        if tx[i:i + preamble_len] == preamble:
            indices.append(i)  
    return indices  
```

```python
with open("tx_bits.bin","rb") as file:
    tx_bits = list(file.read())
    print(tx_bits)
```

```python
rx_indices = find_preamble_all(data, preamble)
tx_indices = find_preamble_all(tx_bits, preamble)
print("Preamble found at indices:", rx_indices)
print("Preamble found at indices:", tx_indices)
```
